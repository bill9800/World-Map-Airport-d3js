# Catch html

```python
import urllib
from bs4 import BeautifulSoup
import urllib.request

def create_soup(url):
    byte_array = urllib.request.urlopen(url)
    content_byte = byte_array.read()
    html = content_byte.decode('utf8')
    byte_array.close()
    return BeautifulSoup(html)
url = input('website')
x = create_soup(url)
print (x)
```
